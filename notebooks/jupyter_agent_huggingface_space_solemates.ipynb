<b>User:</b> From the attached CSV file, count how many shoes are labeled for 'men' (lowercase) and 'women' (lowercase) and plot the distribution as a pie chart.<br>
</div>


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('solemates_shoe_directory.csv') # Assuming the csv file name is solemates_shoe_directory.csv

# Count the number of shoes labeled for 'en' and 'women'
men_count = data['gender'].str.lower().value_counts()['men']
women_count = data['gender'].str.lower().value_counts()['women']

# Plot the distribution as a pie chart
plt.pie([men_count, women_count], labels=['Men', 'Women'], autopct='%1.1f%%')
plt.title('Distribution of Shoes by Gender')
plt.show()
```

The output shows that there are 415 men's shoes and 891 women's shoes in the dataset.

<div class="alert alert-block alert-success">
<b>User:</b> From the attached CSV file ('solemates_shoe_directory.csv'), find the five most common colors in the 'color' (lowercase) column. The 'color' column contains color names as text (e.g., 'black', 'white', 'red'), not hex codes. Display these five colors in a bar chart, with the color names on the x-axis and their count on the y-axis.
</div>


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('solemates_shoe_directory.csv')

# Convert the 'color' column to lowercase for consistency
df['color'] = df['color'].str.lower()

# Count the occurrences of each color
color_counts = df['color'].value_counts()

# Get the top 5 most common colors
top_5_colors = color_counts.nlargest(5)

# Plot the top 5 colors in a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_5_colors.index, top_5_colors.values)
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Top 5 Most Common Colors')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

This code will load the CSV file, count the occurrences of each color, get the top 5 most common colors, and then plot them in a bar chart. The x-axis represents the color names, and the y-axis represents their counts.

<b>User:</b> In the attached CSV file ('solemates_shoe_directory.csv'), analyze the 'heel_type' (lowercase) column, which contains either 'flats' (lowercase) or 'heels' (lowercase) in lowercase. Some rows may have missing values (NaN or None), which should be ignored. Count the number of 'flats' and 'heels' and display the distribution in a pie chart.
</div>


```python
import matplotlib.pyplot as plt
import pandas as pd


# Load the CSV file
df = pd.read_csv('solemates_shoe_directory.csv')


# Ensure the 'heel_type' column is in lowercase for consistency
df['heel_type'] = df['heel_type'].str.lower()


# Count the occurrences of each heel type, ignoring NaN or None values
heel_type_counts = df['heel_type'].value_counts()


# Create a pie chart
plt.pie(heel_type_counts, labels=heel_type_counts.index, autopct='%1.1f%%')
plt.axis('equal') 


plt.title('Distribution of Heel Types')
plt.show()
```
